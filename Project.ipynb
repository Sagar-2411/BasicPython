```python
fig,ax=plt.subplots(figsize=(10,6))
sns.barplot(x='matchType', y='killPoints',ax=ax, data=pubg);
plt.xticks(rotation=70);
#14 -We can see the killpoints are high in 'normal-squad-fpp' and count of 
#this type is also low by which we can conclude that this would be the easiest type!
```

```python
fig,ax=plt.subplots(figsize=(10,6))
sns.barplot(x='matchType', y='weaponsAcquired',ax=ax, data=pubg);
plt.xticks(rotation=70);
#15 -From this graph we can say that there will be more weapons available tha can be equipped in 'normal-solo-fpp' and 'normal-squad-fpp'.
```

```python
#16
cols = pubg.columns
num_cols = pubg._get_numeric_data().columns
num_cols
list(set(cols) - set(num_cols)) #This should give Categorical columns.
#BUT 'Id', 'matchId', 'groupId' has all unique elements. So only 'matchType' is proper Categorical column.
```

```python
fig,ax=plt.subplots(figsize=(13,6))
sns.boxplot(x='matchType', y='winPlacePerc', ax=ax, data=pubg);
plt.xticks(rotation=70);
#17 -From this plot we can say that, in 'normal-duo-fpp' the range is wide and win prediction is highest, too.
#So the conclusion is ,'normal-duo-fpp' can be considered as one of the easy match type. 
```

```python
fig,ax=plt.subplots(figsize=(13,6))
sns.boxplot(x='matchType', y='matchDuration', ax=ax, data=pubg);
plt.xticks(rotation=70);
#18 -From this plot we can say that 'normal-squad-fpp', 'normal-duo-fpp' & 'flaretpp' has
#higher time duration than other modes. So some how these are some complex mode. Or may be 
#in these modes it takes more time to find opponents.
```
